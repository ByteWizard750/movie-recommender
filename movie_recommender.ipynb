```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
import pickle
import gradio as gr
import requests
import os

# Load the movie dataset and similarity matrix
try:
    movies = pickle.load(open('movie_list.pkl', 'rb'))
    similarity = pickle.load(open('similarity.pkl', 'rb'))
    movie_list = movies['title'].tolist()
except Exception as e:
    print(f"Error loading data: {e}")
    # Fallback to a small set of movies if data loading fails
    movies = None
    similarity = None
    movie_list = ["Avatar", "Titanic", "Inception", "The Avengers", "The Matrix"]

# Function to fetch movie poster with error handling
def fetch_poster(movie_id):
    try:
        url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US"
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for 4XX/5XX responses
        data = response.json()
        poster_path = data.get('poster_path')
        if poster_path:
            return "https://image.tmdb.org/t/p/w500/" + poster_path
        else:
            return "https://via.placeholder.com/500x750?text=No+Poster+Available"
    except Exception as e:
        print(f"Error fetching poster for movie ID {movie_id}: {e}")
        return "https://via.placeholder.com/500x750?text=Error+Loading+Poster"

# Function for movie recommendations with error handling
def recommend(movie):
    try:
        if movies is None or similarity is None:
            # Return dummy data if the dataset failed to load
            return (
                ["Error: Dataset not loaded properly"] * 5,
                ["https://via.placeholder.com/500x750?text=Error"] * 5
            )

        # Find the movie index
        movie_indices = movies.index[movies['title'] == movie].tolist()
        if not movie_indices:
            return (
                ["Movie not found in database"] * 5,
                ["https://via.placeholder.com/500x750?text=Movie+Not+Found"] * 5
            )

        index = movie_indices[0]
        # Get similar movies
        distances = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x: x[1])

        recommended_movie_names = []
        recommended_movie_posters = []

        for i in distances[1:6]:  # Get top 5 recommended movies
            try:
                movie_id = movies.iloc[i[0]].movie_id
                recommended_movie_posters.append(fetch_poster(movie_id))
                recommended_movie_names.append(movies.iloc[i[0]].title)
            except Exception as e:
                print(f"Error processing movie at index {i[0]}: {e}")
                recommended_movie_posters.append("https://via.placeholder.com/500x750?text=Error")
                recommended_movie_names.append("Error loading recommendation")

        return recommended_movie_names, recommended_movie_posters
    except Exception as e:
        print(f"Error in recommend function: {e}")
        return (
            [f"Error: {str(e)}"] * 5,
            ["https://via.placeholder.com/500x750?text=Error"] * 5
        )

# Gradio interface function
def gradio_interface(selected_movie):
    if not selected_movie:
        return [], []

    try:
        recommended_movie_names, recommended_movie_posters = recommend(selected_movie)
        return recommended_movie_posters, ", ".join(recommended_movie_names)
    except Exception as e:
        print(f"Error in gradio interface: {e}")
        return (
            ["https://via.placeholder.com/500x750?text=Error"],
            f"Error: {str(e)}"
        )

# Custom CSS for better appearance
css = """
.gradio-container {
    font-family: 'Arial', sans-serif;
}
h1 {
    text-align: center;
    color: #333;
    margin-bottom: 30px;
}
.gallery {
    display: flex;
    flex-wrap: wrap;
    justify-content: center;
}
"""

# Gradio interface setup
with gr.Blocks(css=css) as iface:
    gr.Markdown("# Movie Recommender System")
    with gr.Row():
        with gr.Column(scale=1):
            input_movie = gr.Dropdown(choices=movie_list, label="Select a Movie")
            with gr.Row():
                clear_btn = gr.Button("Clear")
                submit_btn = gr.Button("Submit", variant="primary")

        with gr.Column(scale=2):
            output_gallery = gr.Gallery(label="Recommended Movies", show_label=True, elem_id="gallery").style(grid=5, height="auto")
            output_text = gr.Textbox(label="Movie Names")

    submit_btn.click(fn=gradio_interface, inputs=input_movie, outputs=[output_gallery, output_text])
    clear_btn.click(fn=lambda: [None, ""], outputs=[output_gallery, output_text])

# Launch Gradio interface
iface.launch(share=True)
```

```python
import pandas as pd

# Load the datasets
movies = pd.read_csv("/content/drive/MyDrive/movie_recommend/tmdb_5000_movies.csv")
credits = pd.read_csv("/content/drive/MyDrive/movie_recommend/tmdb_5000_credits.csv")

# Confirm successful load
print("Movies shape:", movies.shape)
print("Credits shape:", credits.shape)

# Merge them
movies = movies.merge(credits, on='title')

# Select useful columns
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]

# Handle missing data
movies.dropna(inplace=True)

import ast

# Extract list of genres, keywords, cast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

def convert_cast(obj):
    L = []
    for i in ast.literal_eval(obj):
        if len(L) < 3:
            L.append(i['name'])
    return L

def fetch_director(obj):
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            return [i['name']]
    return []

# Apply transformations
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(fetch_director)

# Combine all into a single "tags" column
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new_df = movies[['movie_id', 'title', 'tags']]

# Convert tags to string
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

# Vectorize
cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

# Cosine similarity
similarity = cosine_similarity(vectors)

# Save files
pickle.dump(new_df, open('movie_list.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
```
